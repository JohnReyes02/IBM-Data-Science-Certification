### Nube para la ciencia de datos
- Se tomar la información y ponerla en nube
- Evita la limitación física de los computadores
- Permite aprovechar las capacidades de análisis y almacenamiento de equipos avanzados
- Permite implementar algoritmos de computación muy avanzados con la capacidad de hacer cálculos de alto rendimiento.
- Permiten que varias entidades trabajen con los mismos datos al mismo tiempo

- Permite acceso instantáneo a tecnologías de código abierto:
