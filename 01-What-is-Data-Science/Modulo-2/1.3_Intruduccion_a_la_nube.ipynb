## Introducción a la nube

### ¿Qué es el cloud computing?
---
Consiste en la entrega de recursos de computación bajo demanda, como:
- redes
- servidores
- almacenamiento
- aplicaciones
- servicios 
- centros de datos 
a través de internet mediante un sistema de pago por uso.

Aplicaciones y datos que los usuarios acceden en internet y no desde su ordenador local. Algunos ejemplos son:
- Aplicaciones web en línea
- Empleados que utilizan aplicaciones empresariales en línea para realizar su trabajo.
- Almacenamiento de archivos personales
  - OneDrive
  - Google Drive
  - Dropbox

### Beneficios de la computación en la nube
---
- No es necesario comprar e instalar aplicaciones en el computador personal
- Uso de versiones en línea de las aplicaciones y pagos mensuales por suscripción.
- Mayor costo - beneficio
- Acceso a las versiones mas actualizadas del software
- Ahorro en el almacenamiento del computador personal
- Trabajar de forma colaborativa con los mismos archivos en tiempo real

### Computación en la nube
---
Esta compuesto por:
- Cinco características esenciales
- Tres modelos de implementación
- Tres modelos de servicio

### Características de computación en la nube
---
1. Autoservicio bajo demanda.
2. Amplio acceso a la red
  - Acceso a través de teléfonos móviles, tabletas computadores portátiles y estacines de servicio
3. Agrupación de recursos
4. Rápida elasticidad
5. Servicio mesurado

### Modelos de despliegue en la nube
---
1. Nube Pública
2. Nubre Privada
3. Nube Híbirida

### Modelos de servicio en la nube
---
1. IaaS - Infraestructura como servicio
  Se puede acceder a la infraestructura y a los recursos informáticos físicos, sin necesidad de administrarlos ni operarlos.
2. PaaS - Plataforma como servicio
  Puede acceder a la plataforma que incluye las herramientas de hardware y software que normalmente se necesitan para desarrollar e implementar aplicaciones para las usuarios a través de internet
3. SaaS - Software como servicio
  Modelo de licencia y entrega de software en que el software y las aplicaciones se alojan de forma centralizada y se licencian mediante suscripción. (Software a pedido)
